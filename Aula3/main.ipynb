```python
# Analisando 1 cliente
# se o cliente é Bom, OK ou Ruim (Nota de Crédito dele = Score de Crédito)

# Passo a Passo do Projeto
# Passo 0: Entender o desafio da empresa
# Passo 1: Coletar os dados

import pandas as pd
from sklearn.preprocessing import LabelEncoder

dados = pd.read_csv('clientes.csv')

display(dados)
display(dados.info())

# int -> numeros inteiros
# float -> numeros com casa decimal
# object -> string -> textos
```

```python
# Passo 2: Tratar e Preparar a base de dados para a IA

# profissao - advogado, engenheiro, arquiteto...
# mix_credito
# comportamento_pagamento
codificadores = {}

for col in dados.select_dtypes(include=['object']).columns:
    label_encoder = LabelEncoder()
    dados[col] = label_encoder.fit_transform(dados[col])
    codificadores[col] = label_encoder
display(dados)
```
